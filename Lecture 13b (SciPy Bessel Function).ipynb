```python
import numpy as np
import scipy.special as scs
import matplotlib.pyplot as plt
```

**Bessel Function in One Dimension**

```python
# Define Airy Disk equation.
def Bessel(x):
    '''
    Uses the Bessel Function of the first kind to return a normalised intensity I/I_0
    for a position x where x=0 refers to the wave origin.
    '''
    return 4*(scs.j1(x)/x)**2
```

```python
# Create position array.
N = 500
X = np.linspace(-10,10,N)

# Calculate intensity.
I = Bessel(X)

# Plot normalised intensity against position.
plt.semilogy(X,I)
plt.xlabel('x',
           fontsize='14')
plt.ylabel('I / I$_0$',
           fontsize='14')
plt.show()
```

**Bessel Function in Two Dimensions**

```python
# Create position arrays.
N = 500
xmin, xmax = -20, 20
ymin, ymax = -20, 20
X, Y = np.linspace(xmin,xmax,N), np.linspace(ymin,ymax,N)
X, Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)

# Calculate intensity.
I = Bessel(R)

# Plot normalised intensity against radial position.
plt.figure(figsize=(10,5))
plt.imshow(I,
           vmin=0,vmax=0.0010,
           extent = (xmin,xmax,ymin,ymax),
           cmap = 'magma')

plt.xlabel('x',
           fontsize='14')
plt.ylabel('y',
           fontsize='14')
plt.colorbar(label='I / I$_0$')

plt.show
```

```python
# Same plot with log scale.
from matplotlib.colors import LogNorm

plt.figure(figsize=(10,5))
plt.imshow(I,
           norm=LogNorm(),
           extent = (xmin,xmax,ymin,ymax),
           cmap = 'magma')

plt.xlabel('x',
           fontsize='14')
plt.ylabel('y',
           fontsize='14')
plt.colorbar(label='I / I$_0$')

plt.show
```
