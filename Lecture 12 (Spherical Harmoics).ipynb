```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import scipy.special as scs
```

```python
def findtitle(n,m):
    if n == 1:
        dgr = '1st'
    elif n == 2:
        dgr = '2nd'
    elif n == 3:
        dgr = '3rd'
    else:
        dgr = f'{n}th'

    if m == 1:
        odr = '1st'
    elif m == 2:
        odr = '2nd'
    elif m == 3:
        odr = '3rd'
    else:
        odr = f'{m}th'

    return dgr, odr
```

```python
# Create spherical co-ordinate grid.
N = 50
theta, phi = np.linspace(0,2*np.pi,N), np.linspace(0,np.pi,N)
THETA, PHI = np.meshgrid(theta,phi)
```

```python
# Create spherical harmonic data, convert to cartesian and plot.
degree = range(3)
for n in degree:
    order = range(n+1)
    for m in order:
        # Create Spherical Harmonic data.
        SH = scs.sph_harm(m,n,PHI,THETA)
        
        # Convert to Cartesian.
        R = (np.abs(SH), np.real(SH), np.imag(SH))
        axis = ('Absolute', 'Real', 'Imaginary')
        for i in range(3):
            X, Y, Z = R[i]*np.sin(THETA)*np.cos(PHI), R[i]*np.sin(THETA)*np.sin(PHI), R[i]*np.cos(THETA)

            # Plot data.
            dgr, odr = findtitle(n,m)

            if np.sum(Z) != 0.:
                fig, ax = plt.subplots(subplot_kw=(dict(projection='3d')),figsize=(4,4))
                surf = ax.plot_surface(X,Y,Z,
                                       cmap=cm.PuOr,
                                       linewidth=0,
                                       antialiased=True)

                ax.set_xlim(-0.5,0.5)
                ax.set_ylim(-0.5,0.5)
                ax.set_zlim(-0.5,0.5)

                # Add title.
                plt.title(f'Spherical Harmonics ({axis[i]})\n[{dgr} degree, {odr} order]')

                # Add colourbar.
                fig.colorbar(surf, shrink=0.5, aspect=5)

                plt.show()

            else:
                print(f'{dgr} degree, {odr} order has no {axis[i]} part.\n')
            print('______________________________________________________________________\n')
```
