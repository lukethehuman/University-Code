**Task 1:**

Create a grid of points in a (x,y) space with ranges -2 < x < 1; -1.5 < y < 1.5.

The x coordinates will be the real part of the grid of complex numbers, the y coordinates will be the imaginary part of the complex numbers, where c = x+jy.

```python
import numpy as np
N = 501
xmin, xmax = -2.0, +1.0
ymin, ymax = -1.5j, +1.5j
X, Y = np.linspace(xmin,xmax,N), np.linspace(ymin,ymax,N)

C = np.meshgrid(X,Y)
C = C[0]+C[1]
```

**Task 2:**

Implement the Mandelbrot function, using the divergence criteria.

Iterate the values of z up to a maximum of 20 iteratons and record the number of iterations required to reach this threshold.

```python
def mandelbrot(c, n_max = 20):
    n = 0
    z = 0
    while n <= n_max and abs(z) < 4.:
        n += 1
        z = z**2 + c
    if n > n_max:
        n = 4.
    return n
```

```python
iterations = np.zeros((N,N))
for y in range(N):
    for x in range(N):
        iterations[x,y] = mandelbrot(C[x,y])
```

**Task 3:**

Use pyplot.imshow to display the number of iterations taken to diverge with difference colors in the x,y (or Re,Im) plane.

```python
import matplotlib.pyplot as plt

plt.imshow(iterations,
           extent = (xmin,xmax,np.imag(ymin),np.imag(ymax)),
           cmap = 'magma')

plt.title('Mandelbrot Set',fontsize='16')
plt.xlabel('Re',fontsize='14')
plt.ylabel('Im',fontsize='14')

plt.show()
```
