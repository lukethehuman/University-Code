```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define derivative function.
def df(f,x,dx):
    return (f(x+dx)-f(x))/dx
```

```python
# Define Euler Approx.
def Euler_Approx(f,x_0,x_N,dx):
    """
    f = a function of x
    x_0 = initial value
    x_N = final value
    dx = step size
    """
    A = [f(x_0)]
    X = np.arange(x_0,x_N,dx)
    for x in X:
        A.append(f(x)+df(f,x,0.00001)*dx)
    return A
```

```python
# Create exponentially decaying function of time.
def f(t):
    return 2*np.exp(-t/10.5)*np.sin(0.8*t)

# Create time array.
T = np.linspace(0.,50.,200)

# Plot function against time.
plt.figure(figsize=(9,5))
plt.plot(T,f(T),'o-',label='Function')

# Plot Euler approx for the function with various values of dt.
dt_array = (0.1,0.5,1.0,2.5)
for dt in dt_array:
    T = np.arange(0.,50.+dt,dt)
    E = Euler_Approx(f,0.,50.,dt)
    plt.plot(T,E,'.--',label=f'Euler Approx. (dt = {dt})')

plt.xlabel('t')
plt.ylabel('f(t)')

plt.xlim(0.,50.)
plt.grid()
plt.legend()

plt.show()
```
