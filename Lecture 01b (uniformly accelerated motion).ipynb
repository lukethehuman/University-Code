```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initial conditions are:
t_0 = 0 # time = 0 seconds
x_0 = 2 # position = +2 metres
v_0 = -4 # velocity = -4 metres per second
a = 2 # acceleration = +2 metres per second squared
```

```python
# uniformly accelerated motion equations

def findx(x_0,v_0,a,t):
    return(x_0 + v_0 * t + 0.5 * a * t**2)

def findv(v_0,a,t):
    return(v_0 + a * t)
```

```python
# task 1: plot x wrt time t for the first 5 seconds

# set resolution
N = 60

# create time axis
T = np.linspace(0,5,N)

# create corresponding position axis
X = findx(x_0,v_0,a,T)

# create and show plot
plt.plot(T,X)

plt.xlim(0,5)
plt.ylim(-2,6)
plt.axhline(0,color='r',linestyle='--')

plt.xlabel('time (s)',fontsize=14)
plt.ylabel('x (m)',fontsize=14)
plt.title('Uniformly Accelerated Motion (a = -4 m.s$^{-2}$)',fontsize=16)
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# task 2: plot v wrt time t for the first 5 seconds

# initial conditions are:
t_0 = 0 # time = 0 seconds
x_0 = 2 # position = +2 metres
v_0 = -4 # velocity = -4 metres per second
a = 2 # acceleration = +2 metres per second squared

# set resolution
N = 60

# create time axis
T = np.linspace(0,5,N)

# create corresponding position axis
X = findv(v_0,a,T)

# create and show plot
plt.plot(T,X)
plt.xlim(0,5)
plt.ylim(-2,6)
plt.axhline(0,color='r',linestyle='--')

plt.xlabel('time (s)',fontsize=14)
plt.ylabel('v (m/s)',fontsize=14)
plt.title('Uniformly Accelerated Motion (a = -4 m.s$^{-2}$)',fontsize=16)
plt.grid()

plt.show()
```
