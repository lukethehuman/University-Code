```python
import numpy as np
import scipy.integrate as sc_int
import matplotlib.pyplot as plt
```

**Integrating using SciPy.**

```python
# Input data.
x = np.array([0.,1.,2.,3.,4.,4.,6.,7.,8.,9.])
y = np.array([3.6,1.3,0.2,0.9,5.,8.,9.,12.,12.,13.])

# Plot data.
plt.scatter(x,y,
            color='grey',
            label='data')

# Create dataset.
X = sum([list(np.linspace(x[i],x[i+1],18)) for i in range(len(x)-1)],[])
Y = sum([list(np.linspace(y[i],y[i+1],18)) for i in range(len(y)-1)],[])

# Apply polynomial functions and plot.
order = (1,2,3,4,5)
for odr in order:
    p = np.poly1d(np.polyfit(X,Y,odr))
    poly = [p(i) for i in X]
    plt.plot(X,poly,
             label=f'order = {odr}')

# Set plot parameters.
plt.tick_params(direction='in',
                right=True,top=True)
plt.xlabel('x',
           fontsize=14)
plt.ylabel('y',
           fontsize=14)
plt.title('Regression Lines for a Polynomial Function',
          fontsize=16)
plt.legend()

plt.show()
```

```python
# Calculate integrals with SciPy.
datint = sc_int.trapz(y)
plt.axhline(datint,
            label=f'data integral = {datint}')

for odr in order:
    p = np.poly1d(np.polyfit(X,Y,odr))
    poly = [p(i) for i in X]
    quad = sc_int.quad(p,x[0],x[-1])
    plt.scatter(odr,quad[0])
    
# Set plot parameters.
plt.tick_params(direction='in',
                right=True,top=True)
plt.xticks(order)
plt.xlabel('polynomial order',
           fontsize=14)
plt.ylabel('integral result',
           fontsize=14)
plt.title('Integration of a Polynomial Function',
          fontsize=16)
plt.legend()
plt.grid()

plt.show()
```
